### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# # Pyplot Interface (State-Based)
#     Enkel att använda för enkla plottar och enkla anpassningar. 
#     Används för att skapa enkla plottar, diagram och grafer.

# # Object-Oriented Interface (Object-Based)
#     Används för att skapa mer avancerade plottar och anpassningar. 
#     Används för att skapa mer avancerade plottar, diagram och grafer.

# # Rekommenderad
#     Använd object-oriented interface för att skapa plottar. 
#     Det är mer flexibelt och ger dig mer kontroll över dina plottar.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# # Figure
#     En figur är en behållare för alla dina plottar och diagram. 
#     Du kan tänka på en figur som en "ruta" som innehåller alla dina plottar.

# # Axes
#     En axel är en plott inom en figur. 
#     En figur kan innehålla flera axlar. 
#     Du kan tänka på en axel som en "plott" eller "diagram" inom en figur.

# # Axis
#     En axel är en axel inom en axel. 
#     Du kan tänka på en axel som en "x-axel" eller "y-axel" inom en axel.

# # Artist
#     En artist är allt du kan se på plotten. 
#     Det kan vara plottar, axlar, axlar, text, etc.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values and compute y values'
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and an axes
fig, ax = plt.subplots()
ax.plot(x, y, label=r"$y = x^2$")

# Add a title and labels
ax.set_title("Plot of $y = x^2$")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend()

# Show the plot
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# Generate x values and compute y values
x = np.linspace(-4, 4, 100)
y = x**2

# Plot using pyplot
plt.plot(x, y, label=r"$y = x^2$")  # Plot the function

# Add title and labels
plt.title("Plot of $y = x^2$")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()  # Add a legend

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, label="Random points")
ax1.plot(x, y, color="red", label=r"$y = x^2$")
ax1.set_title("Scatter plot")
ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
ax1.legend()

# Bar plot
ax2.bar(names, values, color="skyblue")
ax2.set_title("Bar plot")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Quantity")

# Adjust layout
plt.tight_layout()
plt.show()
```
